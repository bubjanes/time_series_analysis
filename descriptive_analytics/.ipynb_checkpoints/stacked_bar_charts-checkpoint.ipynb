```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline
register_matplotlib_converters()
import seaborn as sns
sns.set_context('notebook') 
sns.set_style('ticks')
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
# importing the data set with FECHA_PAGO as index, parse_date means we are using pandas' datetime function, datefirst adjusts for European date format
df = pd.read_csv('auto_debt.csv', sep=',', encoding='utf-8', index_col='FECHA_PAGO', parse_dates=True, dayfirst=True)
df.head()
```

```python
# put the dates in order
df = df.sort_values('FECHA_PAGO')
```

```python
# add a new column called "Period" with the year-month format
df['Period'] = df.index.to_period(freq='M')
df.head()
```

# Data wrangling: 
### TIPO_ORIGEN = PAO 
### TIPO_MOVIMIENTO = C 
### And six seperate data frames for the six different OBN_CUOTA types: 0, 10, 20, 50, 51, 90

```python
# We only want origen (PAO)
origen_df = df.groupby('MCC_TIPO_ORIGEN')
pao_df = origen_df.get_group('PAO')
pao_df.head()
```

```python
# We only want tipo C (credito)
pao_tipo = pao_df.groupby('MCC_TIPO_MOVIMIENTO')
clean_df = pao_tipo.get_group('C')
clean_df.head()
```

## set up a graphing function

```python
cuota_group = clean_df.groupby('OBN_CUOTA')
```

```python
def cum_sum_graph(self):
    grupo = cuota_group.get_group(self)
    grupo_sum = grupo.groupby('Period').MCC_IMPORTE.sum()
    grupo_cum = grupo_sum.cumsum()
    grupo_cum.plot(kind='bar', title='Cumulative Totals')
```

```python
def mean_series(self):
    grupo_mean = cuota_group.get_group(self)
    print(grupo_mean.groupby('Period').MCC_IMPORTE.mean())
```

## Stacked Bar Chart

```python
indexed = clean_df.set_index('Period')
cuota_group2 = indexed.groupby('OBN_CUOTA')
```

```python
zero = cuota_group2.get_group(0.0)
zero.rename(index=str, columns={"MCC_IMPORTE":"Cuota_0"}, inplace=True)
zero_series = zero.groupby('Period').Cuota_0.sum()
zero_cum = zero_series.cumsum()
zero_df = pd.DataFrame(zero_cum)
zero_df
```

```python
ten = cuota_group2.get_group(10.0)
ten.rename(index=str, columns={"MCC_IMPORTE":"Cuota_10"}, inplace=True)
ten_series = ten.groupby('Period').Cuota_10.sum()
ten_cum = ten_series.cumsum()
ten_df = pd.DataFrame(ten_cum)
ten_df
```

```python
twenty = cuota_group2.get_group(20.0)
twenty.rename(index=str, columns={"MCC_IMPORTE":"Cuota_20"}, inplace=True)
twenty_series = twenty.groupby('Period').Cuota_20.sum()
twenty_cum = twenty_series.cumsum()
twenty_df = pd.DataFrame(twenty_cum)
twenty_df
```

```python
fifty = cuota_group2.get_group(50.0)
fifty.rename(index=str, columns={"MCC_IMPORTE":"Cuota_50"}, inplace=True)
fifty_series = fifty.groupby('Period').Cuota_50.sum()
fifty_cum = fifty_series.cumsum()
fifty_df = pd.DataFrame(fifty_cum)
fifty_df
```

```python
fifty_one = cuota_group2.get_group(51.0)
fifty_one.rename(index=str, columns={"MCC_IMPORTE":"Cuota_51"}, inplace=True)
fifty_one_series = fifty_one.groupby('Period').Cuota_51.sum()
fifty_one_cum = fifty_one_series.cumsum()
fifty_one_df = pd.DataFrame(fifty_one_cum)
fifty_one_df
```

```python
ninety = cuota_group2.get_group(90.0)
ninety.rename(index=str, columns={"MCC_IMPORTE":"Cuota_90"}, inplace=True)
ninety_series = ninety.groupby('Period').Cuota_90.sum()
ninety_cum = ninety_series.cumsum()
ninety_df = pd.DataFrame(ninety_cum)
ninety_df
```

```python
merged = pd.merge(zero_df,ten_df, on='Period', how='outer')
merged_2 = pd.merge(merged,twenty_df, on='Period', how='outer')
merged_3 = pd.merge(merged_2,fifty_df, on='Period', how='outer')
merged_4 = pd.merge(merged_3,fifty_one_df, on='Period', how='outer')
merged_total = pd.merge(merged_4,ninety_df, on='Period', how='outer')
```

```python
new_df = merged_total.sort_index()
new_df
```
